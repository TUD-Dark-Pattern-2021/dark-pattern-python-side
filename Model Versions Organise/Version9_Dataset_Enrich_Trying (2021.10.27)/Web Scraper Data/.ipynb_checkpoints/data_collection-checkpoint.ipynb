```python
import pandas as pd
import numpy as np

# ---- import the raw dataset

raw_1 = pd.read_csv('raw-1.csv')
raw_2 = pd.read_csv('re-taged_normie.csv')

print(raw_1.info())
print(raw_2.info())
```

```python
raw_1
```

```python
raw_2
```

```python
# change the header name of the raw table

raw_1 = raw_1.rename(columns={"content": "Pattern String", "presence": "classification"})

raw_1
```

```python
# Join the two raw dataset

df = pd.concat([raw_1, raw_2])

df.info()
```

```python
 # For later training the model, we should remove the duplicate input to reduce overfitting.

mix_data = df.drop_duplicates(subset="Pattern String")

mix_data.info()
```

```python
mix_data
```

```python
# Remove the rows where the first letter starting with ignoring characters
ignore_str = [',', '.', ';', '{', '}', '#', '/', '?']
filtered_data = mix_data[~mix_data['Pattern String'].str[0].isin(ignore_str)]
```

```python
# Keep the rows where the word count is between 2 and 45
filtered_data = filtered_data[filtered_data['Pattern String'].str.split().str.len() > 1]
filtered_data = filtered_data[filtered_data['Pattern String'].str.split().str.len() < 46]

filtered_data.info()
```

```python
# check the final distribution of the classification after removing the rows with NULL values and duplicates.

print(filtered_data['classification'].value_counts())
```

```python
filtered_data.to_csv('generate-1.csv', index = False, header = True)
```
